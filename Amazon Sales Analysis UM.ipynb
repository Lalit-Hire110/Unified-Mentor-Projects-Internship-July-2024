```python
import pandas as pd 
#import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
df=pd.read_csv("Amazon Sales data.csv")
```

```python
total_Profit=df.Total_Profit.sum()
print('Total Profit:',total_Profit)
total_Revenue=df.Total_Revenue.sum()
print('Total Revenue:',total_Revenue)
```

```python
percentage=[]
for i in range(df.shape[0]):
    pct=(df.Total_Profit[i]/total_Revenue)*100
    percentage.append(round(pct,2))
print(percentage)
df['percentage']=percentage
display(df)
```

```python
plt.figure(figsize=(20, 20))
colors_list = ['Red', 'Green','Orange' ,'Purple','blue','yellow','pink']
graph = plt.bar(df.Region, df.Total_Profit, color=colors_list)
plt.title('Percentage across of all formats')
 
i = 0
for p in graph:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy()
     
    plt.text(x+width/2,
             y+height*1.01,
             str(df.percentage[i])+'%',
             ha='center',
             weight='bold')
    i += 1
plt.show()
```
